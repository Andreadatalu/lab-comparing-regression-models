# 7. Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
def train_and_test_models(models, X_train, y_train, X_test, y_test):
    for model in models:
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)
        
        print("Type of model:", type(model).__name__)
        print("Train_set - Mean Squared Error:", mean_squared_error(y_train, y_train_pred))
        print("Train_set - r2 Score:", r2_score(y_train, y_train_pred))
        print("Test_set - Mean Squared Error:", mean_squared_error(y_test, y_test_pred))
        print("Test_set - r2 Score:", r2_score(y_test, y_test_pred))
        print("________________________________________________")
        print(" ")
```

```python
from sklearn.linear_model import LinearRegression
#from sklearn.linear_model import LogisticRegression (DOESN'T work)
#from sklearn.ensemble import RandomForestClassifier (DOESN'T work)
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

models = [LinearRegression(), RandomForestRegressor(), KNeighborsRegressor()]

train_and_test_models(models, Train_data, y_train, Test_data, y_test)
```
